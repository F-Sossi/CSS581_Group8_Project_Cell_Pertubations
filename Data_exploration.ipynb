```python
import pandas as pd
import numpy as np
import psutil

# Define the path where you want to save the correlation matrices
save_path = 'models'

# Load the dataset with gene expression data (assuming it's in the same format as 'de_train.parquet')
data = pd.read_parquet('data/de_train.parquet')

# Extract gene columns from the data (assuming the first few columns are metadata)
gene_columns = [col for col in data.columns if col not in ['sm_name', 'cell_type']]

# Adjust the batch size to 1000 genes for correlation calculation
batch_size = 1000
num_batches = int(np.ceil(len(gene_columns) / batch_size))

# Function to calculate and save correlation matrices in batches
def calculate_and_save_correlations(data, genes, batch_size, save_path):
    for i in range(num_batches):
        start_index = i * batch_size
        end_index = min((i + 1) * batch_size, len(genes))
        batch_genes = genes[start_index:end_index]
        
        # Calculate the correlation matrix for the batch of genes
        correlation_matrix = data[batch_genes].corr(method='pearson')
        
        # Save the correlation matrix to disk
        file_name = f'correlation_matrix_batch_{i+1}.parquet'
        correlation_matrix.to_parquet(f'{save_path}/{file_name}')
        
        # Print the progress
        print(f"Batch {i+1}/{num_batches} processed and saved.")

# Record the memory usage before processing
memory_before_batches = psutil.virtual_memory().used / (1024 ** 3)

# Calculate and save correlations
calculate_and_save_correlations(data, gene_columns, batch_size, save_path)

# Check memory usage after processing
memory_after_batches = psutil.virtual_memory().used / (1024 ** 3)
additional_memory_used = memory_after_batches - memory_before_batches

print(f"Memory used before batches: {memory_before_batches:.2f} GB")
print(f"Additional memory used: {additional_memory_used:.2f} GB")
print(f"Memory used after batches: {memory_after_batches:.2f} GB")

```

```python
import os
import pandas as pd

# Define the path where you have saved the correlation matrices
save_path = 'models'

# Get a list of all correlation matrix files
correlation_files = [f for f in os.listdir(save_path) if f.startswith('correlation_matrix_batch_')]

# Load the correlation matrices
correlation_matrices = []
for file_name in correlation_files:
    path = os.path.join(save_path, file_name)
    correlation_matrices.append(pd.read_parquet(path))

# If you have enough memory, you can concatenate them into a single DataFrame
# If not, you might need to process them one by one
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Concatenate the correlation matrices if not already done
full_correlation_matrix = pd.concat(correlation_matrices)

# Plot the distribution of correlation values
sns.histplot(full_correlation_matrix.unstack(), bins=50, kde=True)
plt.title('Distribution of Correlation Coefficients')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming `full_correlation_matrix` is a DataFrame where both rows and columns are gene names
# And that you've concatenated your correlation matrices into `full_correlation_matrix`

# Define a threshold for strong correlation
threshold = 0.01

# Find gene pairs with a correlation above the threshold
# Here, we stack the DataFrame to turn it into a Series with a MultiIndex
# Then we filter by our threshold and remove the diagonal (correlation of 1 with itself)
strong_correlations = full_correlation_matrix.stack().loc[lambda x: x.abs() > threshold].loc[lambda x: x != 1]

# Now we can proceed to select a subset of genes for visualization
# As 'strong_correlations' is a Series, we use `.index.get_level_values(0)` to get unique gene names from the first level of the index
subset_genes = strong_correlations.index.get_level_values(0).unique()[:10]

# Ensure that subset_genes are in both rows and columns of the full_correlation_matrix for the .loc indexing to work
heatmap_data = full_correlation_matrix.loc[subset_genes, subset_genes]

# Create a heatmap for the subset
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=0)
plt.title('Heatmap of Gene Correlations')
plt.show()

```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Define the threshold for strong correlations
threshold = 0.8

# Step 2: Create the list of edges from strong_correlations
# Filter the strong correlations to include only those above the threshold
strong_edges = strong_correlations[strong_correlations.abs() > threshold].index.tolist()

# Step 3: Build the network graph
G = nx.Graph()
G.add_edges_from(strong_edges)

# Step 4: Visualize the network
plt.figure(figsize=(40, 40))
pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=7)

# edges
nx.draw_networkx_edges(G, pos, width=1)

# labels
nx.draw_networkx_labels(G, pos, font_size=10, font_family="sans-serif")

plt.axis("off")
plt.show()
```

```python
from pyvis.network import Network

# Assuming 'strong_edges' contains tuples of (node1, node2)
# and 'G' is a networkx graph defined with edges added from 'strong_edges'

# Initialize PyVis network object
net = Network(notebook=True, height="750px", width="100%")

# Add nodes and edges to the PyVis network
net.from_nx(G)

# Customize the network display
net.toggle_physics(True)  # Enable physics for interactive layout
net.show_buttons(filter_=['physics'])  # Show physics buttons to interactively adjust parameters

# Display the network
net.show('gene_network.html')
```
